```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0
    lerning_rate = 0.01
    n = len(x)
    
    plt.scatter(x,y, color='red',marker='*')
    for i in range(2000):        
        y_pred = m_curr*x + b_curr
        loss = (1/n)*sum([val**2 for val in (y - y_pred)])
        plt.plot(x, y_pred, color='green')
        
        md = -(2/n)*sum(x*(y - y_pred))  # mean derivative
        bd = -(2/n)*sum(y - y_pred)      # b derivative
        
        m_curr = m_curr - lerning_rate*md
        b_curr = b_curr - lerning_rate*bd
        
        print("m {}, b {}, i {} and loss {}".format(m_curr, b_curr, i, loss))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
