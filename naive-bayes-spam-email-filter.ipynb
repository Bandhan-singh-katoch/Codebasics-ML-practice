```python
import pandas as pd
```

```python
df = pd.read_csv("../input/spamham-email-data/spam.csv")
df.head(3)
```

```python
dummies = pd.get_dummies(df.Category)
dummies.head()
```

```python
df = pd.concat([df, dummies], axis='columns')
df.head()
```

```python
df = df.drop(['Category','ham'], axis='columns')
df.head()
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
# using sklearn count vectorizer to convert the message text into numerical form
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)

X_test_count = v.transform(X_test.values)

X_train_count.toarray()[:5]
```

# Using naive_bayes MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    "Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat...",
    "Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's",
    "WINNER!! As a valued network customer you have been selected to receivea Â£900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."
]

emails_count = v.transform(emails)

model.predict(emails_count)
```

```python
model.score(X_test_count, y_test)
```

# Using sklearn pipeline to simplify the code

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(emails)
```

```python
clf.score(X_test, y_test)
```
