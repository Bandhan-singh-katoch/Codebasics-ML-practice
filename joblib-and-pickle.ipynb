```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('../input/codebasics-practice/homeprices.csv')
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[["area"]],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

# Save Model to file using pickle

```python
import pickle
```

```python
with open("model_from_pickle",'wb') as file:
    pickle.dump(model,file)
```

```python
# load saved model
with open("model_from_pickle",'rb') as file:
    mp = pickle.load(file)
```

```python
mp.predict([[5000]])
```

# Save trained model using joblib

```python
from joblib import load,dump
```

```python
dump(model,'model_from_joblib')
```

```python
mj = load('model_from_joblib')
```

```python
mj.predict([[5000]])
```
