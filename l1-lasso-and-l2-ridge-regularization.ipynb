```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../input/melbourne-housing-data/Melbourne_housing_FULL.csv")
df.head(3)
```

```python
df.shape
```

**Data modification starts from here**

```python
df.nunique()
```

```python
df_col = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
 
df = df[df_col]
df.head(3)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
col_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[col_zero] = df[col_zero].fillna(0)
df.isna().sum()
```

```python
df['Landsize'] = df.Landsize.fillna(df.Landsize.mean())
df['BuildingArea'] = df.BuildingArea.fillna(df.BuildingArea.mean())
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.shape
```

**One Hot Encoding**

```python
df = pd.get_dummies(df, drop_first=True)
df.head(3)
```

```python
X = df.drop(['Price'], axis='columns')
y = df.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

Here we can see that there is problem of overfitting, now let's try regularization

# **Using Regularization**

**Using L1 Regularizaion or Lasso**

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train,y_train)

lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

**Now Using L2 Regularization or Ridge**

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=100, tol =0.1)

ridge_reg.fit(X_train,y_train)
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
