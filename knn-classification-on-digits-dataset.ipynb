From sklearn.datasets load digits dataset and do following :-

Classify digits (0 to 9) using KNN classifier. You can use different values for k neighbors and need to figure out a value of K that gives you a maximum score. You can manually try different values of K or use gridsearchcv.

Plot confusion matrix

Plot classification report

```python
import pandas as pd

from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3, random_state=2)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Using GridSearchCV to find best value of n_neighbors in knn**

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(KNeighborsClassifier(),{
    'n_neighbors': [3,4,5,6,7,8,9,10]
}, cv=5, return_train_score=False)

clf.fit(X_train, y_train)
```

```python
pd.DataFrame(clf.cv_results_)
```

**From the above table we can analyze that the best n_neighbours = 5 is best(as per GridSearchCV).**

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Now creating the heatmap of the above matrix
import seaborn as sn
from matplotlib import pyplot as plt

plt.figure(figsize=(9,6))
sn.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
